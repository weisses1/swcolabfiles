```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
timeseries1 = r'/content/ts1_clean_Weisse.csv'
ts1_clean = pd.read_csv(timeseries1, parse_dates= ['date']).set_index("date").sort_index()
ts1_clean
```

```python
timeseries2 = r'/content/ts2_clean_Weisse.csv'
ts2_clean = pd.read_csv(timeseries2, parse_dates= ['date']).set_index("date").sort_index()
ts2_clean
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 8), sharex=True)

axes[0].plot(ts1_clean.index, ts1_clean['demand'], color='blue')
axes[0].set_title('Time Series 1 Data')
axes[0].set_xlabel('date')
axes[0].set_ylabel('demand')
axes[0].grid(True)

axes[1].plot(ts2_clean.index, ts2_clean['demand'], color='red')
axes[1].set_title('Time Series 2 Data')
axes[1].set_xlabel('date')
axes[1].set_ylabel('demand')
axes[1].grid(True)

plt.show();
```

```python
columns = ['horsepower', 'curbweight', 'enginesize', 'citympg', 'carwidth']

num = carsData[columns]
```

```python
cat = carsData['fuelsystem']
```

```python
x_dummies = pd.get_dummies(cat, drop_first=True)
```

```python
x_dummies.info()
```

```python
x=pd.concat([num,x_dummies], axis=1)
```

```python
y = carsData['price']
```

```python
# Fit your model
model = LinearRegression().fit(x, y)


# Get R^2
r2 = model.score(x, y)


# n = number of rows, p = number of predictors
n = x.shape[0]
p = x.shape[1]


# Adjusted R^2
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)


print("R^2:", r2)
print("Adjusted R^2:", adj_r2)
```
